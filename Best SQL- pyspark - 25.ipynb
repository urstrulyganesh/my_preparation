### 5. emp earning more than manager

```python
spark.sql(""" 
          select e.emp_name as emp_name,e.salary,
           m.emp_name as maneger_name,m.salary
          from emp_tbl e
          inner join emp_tbl m
          on e.manager_id=m.emp_id
          where e.salary>m.salary
          

          """).display()
```

```python
spark.sql(""" 
          select e.emp_name as emp_name,e.salary,
           m.emp_name as maneger_name,m.salary
          from emp_tbl e
          inner join emp_tbl m
          on e.manager_id=m.emp_id
          where e.salary>m.salary

          """).display()
```

## pyspark

```python
e = emp_df.alias("e")
m = emp_df.alias("m")

# Correct join: employee.manager_id = manager.emp_id
e.join(m, e["manager_id"] == m["emp_id"], "inner") \
 .filter(e["salary"] < m["salary"]) \
 .select(e["emp_name"].alias("emp"), m["emp_name"].alias("manager")) \
 .display()
```

### 21. Emp whose salary < department average

```python
spark.sql("""
    select e.emp_id,
           e.emp_name,
           e.dept_id,
           e.salary,
           d.dept_name
    from emp_tbl e
    join (
        select dept_id, avg(salary) as avg_salary
        from emp_tbl
        group by dept_id
    ) avg_tbl
    on e.dept_id = avg_tbl.dept_id
    join dept_tbl d
    on e.dept_id = d.dept_id
    where e.salary < avg_tbl.avg_salary
""").display()
```

```python
spark.sql("""
    select e.emp_name,
           e.salary,
           dept.dept_name
    from emp_tbl e
    join (
        select dept_id, avg(salary) as avg_salary
        from emp_tbl
        group by dept_id
    ) d
    on e.dept_id = d.dept_id
    join dept_tbl dept
    on e.dept_id = dept.dept_id
    where e.salary < d.avg_salary
""").display()
```

### 24. Case statement

```python
%sql
select 
    *,
    lag(amount) over(order by sale_date) as prev_amount,
    lead(amount) over(order by sale_date) as next_amount,
    case
        when amount > lag(amount) over(order by sale_date) then 'more'
        when amount < lag(amount) over(order by sale_date) then 'less'
        else 'same'
    end as trend
from sales_tbl;
```

```python
w = Window.orderBy("sale_date")

sales_df \
    .withColumn("prev_amount", lag("amount").over(w)) \
    .withColumn("next_amount", lead("amount").over(w)) \
    .withColumn(
        "trend",
        when(col("amount") > lag("amount").over(w), "more")
        .when(col("amount") < lag("amount").over(w), "less")
        .otherwise("same")
    ) \
    .display()
```
